```python
prices=[]
brands=[]
capacity=[]
defrostingtype=[]
ratings=[]
types=[]
builtinstabilizer=[]
warrantyyears=[]
comprehensive=[]  


for i in range(1,20):
    print("""https://www.flipkart.com/search?q=double+door+refrigerator&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i))
    urls="""https://www.flipkart.com/search?q=double+door+refrigerator&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i)
    pages=requests.get(urls)

    data=soup.find_all('div',class_="_30jeq3 _1_WHN1")


for i in data:
    x=soup.find_all('div',class_="_30jeq3 _1_WHN1")
    
    for i in x:
        reg=re.findall('(\d+,\d+)',i.text)
        if len(reg)==1:
            prices.append(reg[0])
        else:
            prices.append(np.nan)   
    

    x=soup.find_all('div',class_="_4rR01T")


    for i in x:
        brands.append(i.text.split()[0])


    for i in x:
        reg=re.findall('\d\d\d',i.text)
        if len(reg)==1:
            capacity.append(reg[0])
        else:
            capacity.append(np.nan)



    for i in x:
        reg=re.findall('[F]\w+\s\w+',i.text)
        if len(reg)==1:
            defrostingtype.append(reg[0])
        else:
            defrostingtype.append(np.nan)




    for i in x:
        reg=re.findall('\s\d\s',i.text)
        if len(reg)==1:
            ratings.append(reg[0])
        else:
            ratings.append(np.nan)



    for i in x:
        reg=re.findall(r'[D]ouble\s\w+',i.text)
        if len(reg)==1:
            types.append(reg[0])
        else:
            types.append(np.nan)

    y=soup.find_all('ul',class_="_1xgFaf")


    for i in y:
        reg=re.findall('Stabilizer(\d)',i.text)
        if len(reg)==1:
            builtinstabilizer.append(reg[0])
        else:
            builtinstabilizer.append(np.nan)


    for i in y:
        reg=re.findall(r'\d\d',i.text)
        if len(reg)==1:
            warrantyyears.append(reg[0])
        else:
            warrantyyears.append(np.nan)




    for i in y:
        reg=re.findall(r'(\d*) Years Warranty on Compressor',i.text)
        if len(reg)==1:
            comprehensive.append(reg[0])
        else:
            comprehensive.append(np.nan)

    
        
```

```python
d={'prices':prices,
  'brands':brands,
  'capacity':capacity,
  'defrostingtype':defrostingtype,
  'ratings':ratings,
  'types':types,
   'builtinstabilizer':builtinstabilizer,
   'warrantyyears':warrantyyears,
   'comprehensive':comprehensive,
  }
```

```python
df=pd.DataFrame(d)
```
